```python
import requests

url = 'https://api.github.com/events'

response = requests.get(url)

def get_first_hundred(in_response):
    response_text = in_response.text
    print(response_text[:100])
    count = 0
    for item in response_text[:100]:
        count += 1
    print(count)

if response.status_code == 200:
    get_first_hundred(response)
elif response.status_code == 404:
    print("Not Found")
else:
    print("Something went wrong", response.status_code)
```

```python
import requests

url = 'https://jsonplaceholder.typicode.com/posts'
params = {'userId': 1, 'id': 2}

response = requests.get(url, params=params)
if response.status_code == 200:
    print(response.url)
    print(response.text)
else:
    print("Something went wrong", response.status_code)
```

Q. Fetch a webpage, check if the request was successful, and extract the main heading text from the page's content.
You can assume that the base url is 'https://example.com'.

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

response = requests.get(url)

if response.status_code == 200:

    if 'text/html' in response.headers.get('Content-Type', ''):

        soup = BeautifulSoup(response.text, 'html.parser')

        main_heading = soup.h1.text

        print(main_heading)
```
