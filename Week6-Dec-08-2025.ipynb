### Polymorphism

```python
class Shape:
    def get_info(self):
        print("Shape")

class Rectangle(Shape):
    def get_info(self):
        print("Rectangle")

class Circle(Shape):
    def get_info(self):
        print("Circle")

circle_1 = Circle()
circle_1.get_info()
```

### Encapsulation

```python
class Dog:
    def __init__(self, name):
        self._name = name

dog_1 = Dog("Tommy")
print(dog_1._name) # X not a good practice (protected attributes and methods usually access inside same class or subclasses only.)
```

Q. Create a python class called StudentRecord that demonstrate strong encapsulation your class must include
1. A private attribute grades that store a list of integers 0-100
2. A method 'add_grades'
    * Accept a grade and ads it's to the list
    * Calls a private method validate grade to verify that a grade is a valid integer within range
3. A method get_average that
    * Calculates and returns the students average score
    * Use a private method "calculate_average" internally

```python
class StudentRecord:
    __grade = []

    def __validate_grade(self, grade):
        if 100 >= grade >= 0:
            return True
        return False

    def add_grade(self,grade):
        print(self.__validate_grade(grade))
        if self.__validate_grade(grade):
            self.__grade.append(grade)
        return None

    def __calculate_average(self):
        print(sum(self.__grade))
        print(len(self.__grade))
        return sum(self.__grade) / len(self.__grade)

    def get_average(self):
        return self.__calculate_average()



record_1 = StudentRecord()
record_1.add_grade(60)
record_1.add_grade(-40)
record_1.add_grade(50)
record_1.get_average()
```

Q2. Write a python code that reads a file called input.txt and write only even-numbers lines into a new file called "even_lines.txt" (consider the first line as the line number one)

```python
with open("./file_handling/index.txt", "r") as input_file:
    content = input_file.readline()
    print(content.strip())
```

```python

with open("index.txt", "r") as infile:
    lines = infile.readlines()

with open("even_lines.txt", "w") as outfile:
    for line_number, line in enumerate(lines, start=1):
        if line_number % 2 == 0:
            outfile.write(line)
```
