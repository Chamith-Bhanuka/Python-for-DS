Q1. Write a python class name 'Car' with the following specifications
* A class attribute 'wheels' that stores a number of wheels for car
* An instance method drive_to that prints 'Driving to |destination| on |number_of_wheels| wheels'
* Another instance method that prints 'This car runs on |number_of_wheels| wheels'

```python
class Car:
    wheels = 4

    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def drive_to(self, destination):
        return f"{self.color} {self.brand} driving to {destination}"

    def get_wheels(self):
        return f"{self.color} {self.brand} car runs on {self.wheels} wheels"

car_1 = Car("BMW", "Black")
print(car_1.drive_to("Galle"))
print(car_1.get_wheels())
```

Q2. Create a python class name Circle that represents a circle. your class should be able to
1. Store the radius of the circle
2. Calculate and return the area of the circle
3. Calculates and returns the circumference of circle
4. Include a method display info. that prints the radius, area, and circumference of the circle in a clear readable format

```python
import math

class Circle:
    PI_VALUE = math.pi

    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def calculate_circumference(self):
        return 2 * math.pi * self.radius

    def get_info(self):
        return f"radius: {self.radius} cm \narea: {self.calculate_area()} cm^2 \ncircumference: {self.calculate_circumference()} cm"

circle = Circle(7)
print(circle.get_info())
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} Animal speaking..!")

class Dog(Animal):
    def __init__(self, color, name):
        super().__init__(name)
        self.color = color

    def speak(self):
        print(f"{self.name}Dog barks..!")

dog_1 = Dog("Black", "Fluffy")
dog_1.speak()
print(dog_1.color)
print(dog_1.name)
```
