Q1. Extract all elements that are
   1. Grater than 10 or less than 5
   2. And not equal to 15

```python
import numpy as np

arr = np.array([
    [12,5,7,20],
    [3,15,8,2],
    [10,25,6,18],
    [4,11,30,9]
])

print(arr[((arr > 10) | (arr < 5)) & (arr != 15)])
```
