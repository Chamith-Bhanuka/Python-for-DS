Q1. Extract all elements that are
   1. Grater than 10 or less than 5
   2. And not equal to 15

```python
import numpy as np

arr = np.array([
    [12,5,7,20],
    [3,15,8,2],
    [10,25,6,18],
    [4,11,30,9]
])

print(arr[((arr > 10) | (arr < 5)) & (arr != 15)])
```

```python
import pandas as pd

df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold": [12, 50, 34, 20, 40],
    "Price": [120000, 1500, 3500, 25000, 6000],
    "Discounts": [0.10, 0.05, 0.08, 0.12, 0.07]
})
print(df)
```

Q1.  1. Calculate the final price after discount for each product
     2. Calculate total revenue for each product
        (for above part 1 and part 2 add both result as new columns to the data frame)
     3. Filter and display only the products where total revenue is grater than 500000
     4. Calculate the total revenue of all products combine

```python
# 01
df["FinalPrice"] = df["Price"] * (1 - df["Discounts"])

# 02
df["TotalRevenue"] = df["FinalPrice"] * df["UnitSold"]
print(df)

# 03
high_revenue = df[df["TotalRevenue"] > 500000]
print()
print(high_revenue)

# 04
total_revenue_all = df["TotalRevenue"].sum()
print()
print(total_revenue_all)
```
